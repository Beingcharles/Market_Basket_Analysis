**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

**Import Data to notebook or file**

```python
mba=pd.read_csv('/content/MBA dataset.csv',header=None)
```

```python
mba.head(25)
```

**Convertion of Data into List Format**

> For Aprori Algorithm, we need to convert data into list format



```python
transactions=[]
for i in range (0,7501):
  transactions.append([str(mba.values[i,j])for j in range(0,20)])
```

**Training the Apriori Model on our Dataset**

```python
Apriori_Rules=apriori(transactions=transactions,
                      min_support=0.003,
                      min_confidence=0.2,
                      min_lift=3,
                      min_length=2,
                      max_length=2)
```

```python
Results=list(Apriori_Rules)
```

```python
Results
```

**Putting the results in a well organised into Pandas Dataframe**

```python
def inspect(results):
  lhs= [tuple(result[2][0][0])[0] for result in results]
  rhs= [tuple(result[2][0][1])[0] for result in results]
  support =[result[1] for result in results]
  confidences= [result[2][0][2] for result in results]
  lifts= [result[2][0][3] for result in results]
  return list (zip(lhs, rhs, support,confidences, lifts))

df= pd.DataFrame(inspect(Results),columns=['Left Hand Side','Right Hand Side','Support','confidence','Lift'])
```
